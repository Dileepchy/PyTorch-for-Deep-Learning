```python
#Importing Numpy Library
import numpy as np
#Generating 200 values of x in between [-10,10]
x = np.linspace(-10, 10, 200)
print(x)
print(x.shape) # Validating the number of vlaues of x
```

```python
#Defining Sigmoid Function
def sigmoid(x):
    return [(1/(1+np.exp(-i))) for i in x]

#Defining Hyperbolic Tan Function
def tanh(x):
    return [((1-np.exp(-2*i))/(1+np.exp(-2*i))) for i in x]
```

```python
#Passing values of x to sigmoid function
y_sig=sigmoid(x)

#Passing values of x to tanh function
y_tanh=tanh(x)
```

```python
#Importing Matplotlib Library
import matplotlib.pyplot as plt
%matplotlib inline

#Scatter Plot for Sigmoid Function
plt.scatter(x,y_sig, c='b', label = 'Sigmoid')
#Scatter Plot for Tanh Function
plt.scatter(x,y_tanh, c="r", label = 'tanh')
plt.legend(loc='lower right')
#Labels
plt.title('Scatter plot for Sigmoid and Tanh Function')
plt.xlabel('input to activations(x)')
plt.ylabel('output values(y)')
plt.show()
```

### Variants of ReLU

```python
#Defining Function for ReLU and its variants
def Relu(x):
    return [ (lambda i: i if (i>0) else 0)(i) for i in x]

def leaky_relu(x):
    return [ (lambda i: i if (i>=0) else 0.1*i)(i) for i in x]

def sin_relu(x):
    epsilon = 0.1
    return [ (lambda i: i if (i>0) else epsilon*(np.sin(i)-np.cos(i)))(i) for i in x]

def rand_relu(x):
    # np.random.seed(1)
    alpha = np.random.rand(1)
    return [ (lambda i: i if (i>0) else alpha* i)(i) for i in x]
```

```python
y_relu = Relu(x)
y_lrelu = leaky_relu(x)
y_srelu = sin_relu(x)
y_rrelu = rand_relu(x)

#ReLU=RED
plt.scatter(x, y_relu, c='r',label='ReLU')
#Leaky ReLU=GREEN
plt.scatter(x, y_lrelu, c='g', label = 'Leaky ReLU')
#SReLU=Blue
plt.scatter(x, y_srelu, c='b', label = 'SReLU')
#RReLU=Yellow
plt.scatter(x, y_rrelu, c='y', label = 'RReLU')
plt.legend(loc='lower right')

#Visualizing in a combine plot
plt.title('Scatter plot for ReLU, Leaky ReLU, SReLU and RReLU')
plt.xlabel('input to activations(x)')
plt.ylabel('output values(y)')
plt.show()
```

```python
#Visualizations in multiple subplots
plt.figure(figsize=(10,10))
plt.subplot(321)
plt.tight_layout()
plt.scatter(x, y_relu, c='r')
plt.title('RELU')
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(322)
plt.tight_layout()
plt.scatter(x, y_lrelu, c='g')
plt.title('Leaky Relu')
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(323)
plt.tight_layout()
plt.scatter(x, y_srelu, c='b')
plt.title('Sin Relu')
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(324)
plt.tight_layout()
plt.scatter(x, y_rrelu, c='y')
plt.title("Randomized Relu")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
